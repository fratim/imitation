```python
import torch
import torch.nn as nn
import torch.nn.functional as F


class Net(torch.nn.Module):

    def __init__(self):
        super(Net, self).__init__()
        self.fc1 = nn.Linear(2,2, bias=False)  # 5*5 from image dimensio

    def forward(self, x):
        x = self.fc1(x)
        return x


net = Net()
print(net)
```

```python
X = torch.rand(100,2)
y = torch.concat(([X[:,(1,)],X[:,(0,)]]), dim=1)
```

```python
a = torch.rand(1)*256
a = a.long()
X[a,:]
```

```python
import torch.optim as optim

criterion = nn.MSELoss()

# create your optimizer
optimizer = optim.SGD(net.parameters(), lr=0.0001)
samples = 10000
for _ in range(samples):
    rand_row = a = torch.rand(1)*X.shape[0]
    rand_row = rand_row.long()
    optimizer.zero_grad()   # zero the gradient buffers
    output = net(X[rand_row:rand_row+5, :])
    loss = criterion(output, y[rand_row:rand_row+5, :])
    print(loss)
    loss.backward()
    optimizer.step()
```
